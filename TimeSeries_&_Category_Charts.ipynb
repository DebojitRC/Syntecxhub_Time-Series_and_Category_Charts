```python
## Objective

# To analyze sales data using time series and category-based visualizations.
# This includes line charts for trends, bar charts for comparison, and pie charts for proportion analysis.
```

```python
## Import Libraries

# Pandas is used for data handling, while Matplotlib is used for creating visualizations.
```

```python
# Import Libraries

import pandas as pd
import matplotlib.pyplot as plt
```

```python
## Load Dataset

# The dataset is loaded from a CSV file into a DataFrame for analysis.
```

```python
# Load Dataset

df = pd.read_csv("Superstore Dataset.csv", encoding='latin1')
df.head()
```

```python
## Inspect Dataset

# This step helps understand the dataset structure, including column names and data types.
```

```python
# Inspect Dataset

df.columns
```

```python
## Data Preparation

# The 'Order Date' column is converted into datetime format to enable time-based analysis.
# A new column is created for monthly aggregation.
```

```python
# Data Preparation

df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Month'] = df['Order Date'].dt.to_period('M')

df.head()
```

```python
## Time Series Analysis (Line Chart)

# Sales are aggregated monthly to observe trends over time.
# A line chart is used because it effectively shows changes over a continuous time period.
```

```python
# Time Series Analysis (LIne Chart)

monthly_sales = df.groupby('Month')['Sales'].sum()

monthly_sales.plot(figsize=(10,5))

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid()

plt.savefig("monthly_sales.png")
plt.show()
```

```python
## Category Comparison (Bar Chart)

# A bar chart is used to compare total sales across different categories, making it easy to identify which category performs best.
```

```python
# Category Comparison (Bar Chart)

category_sales = df.groupby('Category')['Sales'].sum()

category_sales.plot(kind='bar', figsize=(8,5))

plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales")

plt.savefig("category_sales.png")
plt.show()
```

```python
## Category Distribution (Pie Chart)

# A pie chart is used to show the proportion of total sales contributed by each category.
```

```python
# Category Distribution (Pie Chart)

category_sales.plot(kind='pie', autopct='%1.1f%%', figsize=(6,6))

plt.title("Category Sales Distribution")
plt.ylabel("")

plt.savefig("category_pie.png")
plt.show()
```

```python
## Insights

# The visualizations help identify key patterns and trends in the data.
# Monthly sales show trends and fluctuations over time
# Certain categories contribute more to overall sales
# There is a clear difference in performance across categories
```
